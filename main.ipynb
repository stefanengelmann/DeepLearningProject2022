```python
#Importing packages: 
#import matplotlib.pyplot as plt
import numpy as np
import json    
import IPython
import math
import matplotlib.pyplot as plt
```

```python
#Loading data from log file: 
file = open("logs/final_log.json")
data = json.load(file)
valid_array = []
train_array = []
for i in range(len(data)):
    valid_array.append(data[i]["valid"])
    train_array.append(data[i]["train"])
print(min(valid_array))
```

```python
plt.figure()
x_range = np.arange(0,len(valid_array))
plt.plot(x_range,train_array)
plt.plot(x_range,valid_array)
plt.legend(("Train Loss", "Valid Loss"))
plt.show()
```
